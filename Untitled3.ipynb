```python
df=df.drop(["Like"],axis=1)
```

```python
df.head()
```

```python
df1=df.iloc[:,:-4]
```

```python
df1
```

```python
from sklearn.decomposition import PCA
from sklearn import preprocessing
pca_data = preprocessing.scale(df1)

pca = PCA(n_components=11)
pc = pca.fit_transform(df1)
names = ['pc1','pc2','pc3','pc4','pc5','pc6','pc7','pc8','pc9','pc10','pc11']
pf = pd.DataFrame(data = pc, columns = names)
pf
```

```python
loadings = pca.components_
num_pc = pca.n_features_
pc_list = ["PC"+str(i) for i in list(range(1, num_pc+1))]
loadings_df = pd.DataFrame.from_dict(dict(zip(pc_list, loadings)))
loadings_df['variable'] = df1.columns.values
loadings_df = loadings_df.set_index('variable')
loadings_df
```

```python
#Correlation matrix plot for loadings 
plt.rcParams['figure.figsize'] = (20,15)
ax = sns.heatmap(loadings_df, annot=True, cmap='rocket')
plt.show()
```

```python
from sklearn.cluster import KMeans
wcss = []
for k in range(1,10):
    kmeans = KMeans(n_clusters=k, init="k-means++")
    kmeans.fit(df1)
    wcss.append(kmeans.inertia_)
```

```python
plt.figure(figsize=(12,6))    
plt.grid(True)
plt.plot(wcss, linewidth=2, color="red", marker ="8")
plt.xlabel("K-value")
plt.ylabel("WCSS")
```

```python
km = KMeans(n_clusters=4)
clusters = km.fit_predict(df1)
```

```python
clusters
```

```python
df["label"]=clusters
```

```python
df
```
